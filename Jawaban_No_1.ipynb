1. Lakukan sebuah regresi dengan menggunakan Algoritma KNN dalam soal No. 1, dengan ketentuan sebagai berikut :

a. Apabila Cuaca buruk dengan nilai = 1, Weekday, dan Game = 0, maka berapa roti yang harus dibuat?

b. Apabila Cuaca baik dengan nilai 4, Weekend, dan Game =1, maka berapa roti yang harus dibuat?

```python
import pandas as pd
import numpy as np
from pydrive.auth import GoogleAuth
from sklearn.neighbors import KNeighborsRegressor
```

```python
store_data = pd.read_excel('/content/drive/MyDrive/Classroom/Konsep Aplikasi Datamining A Genap 2023-2024/UAS/dataset soal no.1.xls')
store_data
```

```python
df = store_data.copy()
df
```

```python
# Pendefinisian
X = df[['Weather', 'Holiday', 'Game']]
y = df['Qty']

# Data Model
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X, y)

# Skenario
scenario_1 = pd.DataFrame([[1, 0, 0]], columns=['Weather', 'Holiday', 'Game'])
scenario_2 = pd.DataFrame([[4, 1, 1]], columns=['Weather', 'Holiday', 'Game'])

# Prediksi Skenario
prediction_1 = knn.predict(scenario_1)
prediction_2 = knn.predict(scenario_2)

prediction_1, prediction_2
```

```python
print('a. Apabila Cuaca buruk dengan nilai = 1, Weekday, dan Game = 0, maka jumlah roti yang harus dibuat adalah sekitar', prediction_1[0])
print('b. Apabila Cuaca baik dengan nilai = 4, Weekend, dan Game = 1, maka jumlah roti yang harus dibuat adalah sekitar', prediction_2[0])
```
